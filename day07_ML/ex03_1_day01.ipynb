#### csv 파일을 DataFrame type으로 loading하기

```python
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv') # csv파일이 DataFrame으로 변환됨
print('train_df type :',type(train_df))
```

#### DataFrame 생성하기

```python
dict1 = {'Name':['이순신','유관순','안중근','윤봉길'],
         'Year':[2001,2002,2003,2004],
         'Gender':['Male','Female','Male','Male']}

# Dictionary 를 DataFrame 으로 변환하기
data_df = pd.DataFrame(dict1)
print(type(data_df))
print(data_df)
print('-'*60)

# 새로운 column 명 추가하기
data_df1 = pd.DataFrame(dict1, columns=['Name','Year','Gender','Age'])
print(data_df1)
print('-'*60)

# index 를 새로운 값으로 할당하기
data_df2 = pd.DataFrame(dict1, index=['one','two','three','four'])
print(data_df2)
print('-'*60)
```

#### DataFrame : 컬럼명, 인덱스

```python
print('columns :',train_df.columns)
print('index :',train_df.index)
print('index value :',train_df.index.values)
```

```python
# DataFrame 객체에서 []연산자 속에 한 개의 컬럼만 (문자열로) 입력하면 Seriese 객체를 반환함
series = train_df['Name']
print(series.head(3))
print(type(series))
print('-'*60)

# DataFrame 객체에서 []연산자 속에 한 개의 컬럼만 (문자열로) 리스트로 입력하면 DataFrame 객체를 반환함
series2 = train_df[['Name']]
print(series2.head(3))
print(type(series2))
print('-'*60)

# DataFrame 객체에서 []연산자 속에 여러 개의 컬럼을 (문자열로) 리스트로 입력하면 DataFrame 객체를 반환함
extract_df = train_df[['Name', 'Age']]
print(extract_df.head(3))
print(type(extract_df))
```

```python
#  DataFrame은 행(row)과 열(column)을 가지고 있는 data type임
print('train_df DataFrame 의 크기 :',train_df.shape) # (891, 12)
```

```python
# info()
# DataFrame 의 컬럼명, 데이터타입, 
#             NUll 건수, data 건수 를 알려줌
train_df.info()
```

```python
# describe()
# data 값들의 평균, 표준편차, 4분위 분포도를 알려줌
# 숫자형 컬럼들에 대해서 해당 정보를 알려줌
train_df.describe()
```

```python
# value_counts()
# 동일한 개별 데이터 값이 몇 건이 있는지 정보를 알려줌 
# <-- 개별 데이터값의 분포도를 알려줌
# value_counts() 는 Seriese 객체에서만 사용할 수 있음 
# <-- DataFrame을 단일 컬럼으로 입력해서 Seriese로 변환한 후에 사용해야 함


seriese_pclass = train_df['Pclass']
print(type(seriese_pclass))
print('-'*30)
print(seriese_pclass)
print('-'*30)

print(seriese_pclass.head())
print('-'*30)

value_counts_result = train_df['Pclass'].value_counts()
print(value_counts_result)
print('-'*30)
```

```python
# sort_values() by=정렬할컬럼, ascending=(True)False 
# (기본값(True-오름차순), False는 내림차순)
result1 = train_df.sort_values(by='Pclass', ascending=False)
print(result1)
print('-'*60)

result2 = train_df[['Name','Age']].sort_values(by='Age')
print(result2)
print('-'*60)

result3 = train_df[['Name','Age','Pclass']].sort_values(by=['Pclass','Age'])
print(result3)
print('-'*60)
```
