### DataFrame, list, dictionary, ndarray(numpy)

#### list, ndarray(numpy) 에서 DataFrame으로 변환하기

```python
import pandas as pd
import numpy as np

col_name1 = ['col1']
list1 = [1, 2, 3]
arr1 = np.array(list1)

print('arr1 shape :',arr1.shape)
print('list1 type :', type(list1))
print('arr1 type :', type(arr1))
print('-'*60)

df_list1 = pd.DataFrame(list1, columns=col_name1)
print('df_list1 - 1차원 리스트로부터 생성한 DataFrame :\n',df_list1)
print('df_list1 shape :',df_list1.shape)
print('-'*60)

df_arr1 = pd.DataFrame(arr1, columns=col_name1)
print('df_arr1  - 1차원 ndarray부터 생성한 DataFrame :\n',df_arr1)
print('df_arr1 shape :',df_arr1.shape)
print('-'*60)
```

```python
# 컬럼이름으로 사용할 list(col_name2)
col_name2 = ['col1','col2','col3']

# 2행 X 3열 형태의  list 와 ndarray를 생성하고
# 이 list와 ndarray를 DataFrame으로 변환하기
list2 = [[1, 2, 3],
         [11, 22, 33]]
arr2 = np.array(list2)

print('arr2 shape :', arr2.shape)
print('-'*60)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('df_list2 - 2차원 리스트로부터 생성한 DataFrame :\n',df_list2)
print('df_list2 shape :',df_list2.shape)
print('-'*60)

df_arr2 = pd.DataFrame(arr2, columns=col_name2)
print('df_arr2 - 2차원 ndarray로부터 생성한 DataFrame :\n',df_arr2)
print('df_arr2 shape :',df_arr2.shape)
print('-'*60)

```

#### dictionary를 DataFrame으로 변환하기

```python
# key는 컬럼명으로 매핑, Value는 list나 ndarray로 지정함
dict_data = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict_data = pd.DataFrame(dict_data)
print('dictionary 로부터 생성된 DataFrame :\n', df_dict_data)
```

#### DataFrame을 ndarray 로 변환하기

```python
arr3 = df_dict_data.values
print('arr3 :\n',arr3)
print('arr3 type :',type(arr3))
print('arr3 shape :',arr3.shape)
```

#### DataFrame(df_dict_data)을 list 와 dictionary 로 변환하기

```python
# list 로 변환
list_df = df_dict_data.values.tolist()
print('list_df :\n',list_df)
print('list_df type :',type(list_df))
print('-'*60)


# dictionary로 변환
dict_df1 = df_dict_data.to_dict()
print('dict_df1 :\n',dict_df1)
print('dict_df1 type :',type(dict_df1))
print('-'*60)

dict_df2 = df_dict_data.to_dict('list')
print('dict_df2 :\n',dict_df2)
print('dict_df2 type :',type(dict_df2))
print('-'*60)
```

### DataFrame 의 컬럼 데이터셋 Access 
* DataFrame의 컬럼 데이터셋의 생성과 수정은 []연산자를 사용해서 할 수 있음
* 새로운 컬럼에 값을 할당하려면 DataFrame[] 내에 새로운 컬럼명을 입력하고 값을 할당함 

```python
train_df = pd.read_csv('train.csv') # csv파일이 DataFrame으로 변환됨
print('train_df type :',type(train_df))
```

```python
# del train_df['Age_0'] <-- 컬럼 삭제하기

train_df.head()
```

```python
# 컬럼 추가하기
train_df['Age_0'] = 0
train_df.head()
```

```python
train_df['Age_by_10'] = 0
train_df.head()
```

```python
train_df['Family_No'] = train_df['SibSp'] + train_df['Parch']
train_df.head()
```

* 기존 컬럼에 있는 값을 업데이트하려면

### index  객체

```python
# 원본 파일 다시 로딩하기
train_df = pd.read_csv('train.csv')

# index 추출하기
indexes = train_df.index
print(indexes)

# index 객체를 array로 변환하기
print('index 객체 array :\n',indexes.values)
```

```python
print(type(indexes.values))
print(indexes.values.shape)
print(indexes[:5].values)
print(indexes.values[:5])
print(indexes[6])
```

```python
# Index does not support mutable operations
# index 객체에는 [] 연산자를 사용해서 값을 변경할 수 없음
# indexes[0] = 33
```

```python
series_fare = train_df['Fare']
series_fare.head()
```

```python
print('Fare Series max 값 :',series_fare.max())
print('Fare Series sum 값 :',series_fare.sum())
print('sum(series_fare)   :',sum(series_fare))

# 각 row마다 3이 더해짐
# index는 data를 구별하는 역할만 하고
# 실제 연산이 이루어질 때 연산되어지지는 않음
print((series_fare + 3).head())
print('-'*60)
series_fare.head()
```
